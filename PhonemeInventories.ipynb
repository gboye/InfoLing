```python
import pandas as pd
import codecs
```

```python
nomBoth="/Users/gilles/Desktop/Leah-PhonemeInventoriesBOTH.csv"
nomEth="/Users/gilles/Desktop/Leah-PhonemeInventoriesETH.csv"
nomTheta="/Users/gilles/Desktop/Leah-PhonemeInventoriesTHETA.csv"
```

```python
def openInventory(nomInventory):
    with codecs.open(nomInventory,"r",encoding="utf8") as input:
        lines=input.readlines()
    return lines[1:]
```

```python
linesBoth=openInventory(nomBoth)
linesEth=openInventory(nomEth)
linesTheta=openInventory(nomTheta)
```

```python
def setPhonemes(lines):
    phonemes=set()
    for line in lines:
        items=line.strip().split("\t")[2:]
        for item in items:
            phonemes.add(item)
    return phonemes
```

```python
bothPhonemes=setPhonemes(linesBoth)
ethPhonemes=setPhonemes(linesEth)
thetaPhonemes=setPhonemes(linesTheta)
allPhonemes=bothPhonemes|ethPhonemes|thetaPhonemes
```

```python
phonemes=sorted(list(allPhonemes))
for phoneme in phonemes:
    print u"%s,"%phoneme,
```

```python
def STDLines(lines,phonemes,title=["Language","Source"],eththeta=[]):
    phonemes=sorted(list(phonemes))
    sortieLines=[]
    if title:
        sortieLines.append(title+phonemes)
    for line in lines:
        language,source=line.strip().split("\t")[:2]
        inventory=line.strip().split("\t")[2:]
        phonemesRow=[]
        for phoneme in phonemes:
            if phoneme in inventory:
                phonemesRow.append("1")
            else:
                phonemesRow.append("0")
        sortieLines.append([language,source]+eththeta+phonemesRow)
    return sortieLines
```

```python
sortieLines=STDLines(linesBoth,bothPhonemes)
with codecs.open(nomBoth.replace(".csv","-STD.csv"),"w",encoding="utf8") as output:
    for line in sortieLines:
        output.write("\t".join(line)+"\n")
```

```python
sortieLines=STDLines(linesEth,ethPhonemes)
with codecs.open(nomEth.replace(".csv","-STD.csv"),"w",encoding="utf8") as output:
    for line in sortieLines:
        output.write("\t".join(line)+"\n")
```

```python
sortieLines=STDLines(linesTheta,thetaPhonemes)
with codecs.open(nomTheta.replace(".csv","-STD.csv"),"w",encoding="utf8") as output:
    for line in sortieLines:
        output.write("\t".join(line)+"\n")
```

```python
sortieLines=[]
sortieLines.extend(STDLines(linesBoth,allPhonemes,["Language","Source",u"eth",u"theta"],["1","1"]))
sortieLines.extend(STDLines(linesEth,allPhonemes,[],["1","0"]))
sortieLines.extend(STDLines(linesTheta,allPhonemes,[],["0","1"]))
with codecs.open(nomBoth.replace("BOTH.csv","ALL-STD.csv"),"w",encoding="utf8") as output:
    for line in sortieLines:
        output.write("\t".join(line)+"\n")
```
