```python
import urllib2
from bs4 import BeautifulSoup
```

```python
listeSingers=["John_Lennon","Paul_McCartney","Ringo_Star"]
```

```python
listeBios=[]
for nom in listeSingers:
    response = urllib2.urlopen('http://en.wikipedia.org/wiki/%s' % nom)
    html = response.read()
    soup=BeautifulSoup(html)
    listeBios.append(list(soup.find("table").next_siblings)[1].text)
```

```python
from pattern.en import parse
from pattern.en import pprint
```

```python
for bio in listeBios:
    pprint(parse(bio, relations=True, lemmata=True))
```
