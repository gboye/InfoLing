```python
import codecs
from lxml import etree
from lxml.builder import E
import random, re
```

```python
debug=True
```

```python
score={0:"0", 1:"100", 2:"50", 3:"33.333", 4:"25", 5:"20"}

class MoodleXML:
    def __init__(self):
        self.header=[u'<?xml version="1.0" encoding="UTF-8"?>',u'<quiz>']
        self.trailer=[u'</quiz>']
        self.quizzes=[]
    def addQuiz(self,quiz):
        self.quizzes.append(quiz)
    def addQuizzes(self,quizzes):
        for quiz in quizzes:
            self.addQuiz(quiz)
    def getXML(self):
        result=u"\n".join(self.header)+u"\n"+u"\n".join(self.quizzes)+u"\n"+u"\n".join(self.trailer)
        return result

class XMLMultichoice:
    '''
    Conteneur pour une série d'exercices d'une même catégorie
    '''
    def __init__(self,category):
        self.category=category
        self.exercices=[]
    
    def addExercice(self,exercice):
        self.exercices.append(exercice)
    
    def getMultichoices(self):                
        categoryStructure=[u'<question type="category">',
                            u'<category>',
                                u'<text>',
                                    self.category,
                                u'</text>',
                            u'</category>',
                        u'</question>'
                        ]
        result=u"\n".join(categoryStructure)
        for exercice in self.exercices:
            result+=exercice.forme
        return result

    
    
class MultichoiceExercice:
    '''
    Conteneur pour un exercice Multichoice
    
    on fournit le titre et le corps de la question à insérer tout prêts
    '''
    def __init__(self,titre,consigne,reponses):
        self.titre=titre
        self.consigne=consigne
        self.generalFeedback=u"Bien reçu."
        exerciceHeader=[ 
            u'<question type="multichoice">',
                u'<name><text>%s</text></name>'%self.titre,
                u'<questiontext><text><![CDATA[%s]]></text></questiontext>'%self.consigne,
                u'<generalfeedback format="html">',
                  u'<text><![CDATA[<p>%s<br/></p>]]></text>'%self.generalFeedback,
                u'</generalfeedback>',
                u'<defaultgrade>1.0000000</defaultgrade>',
                u'<penalty>0.3333333</penalty>',
                u'<hidden>0</hidden>',
                u'<single>false</single>',
                u'<shuffleanswers>true</shuffleanswers>',
                u'<answernumbering>123</answernumbering>',
                u'<correctfeedback format="html">',
                  u'<text>Votre réponse est correcte.</text>',
                u'</correctfeedback>',
                u'<partiallycorrectfeedback format="html">',
                  u'<text>Votre réponse est partiellement correcte.</text>',
                u'</partiallycorrectfeedback>',
                u'<incorrectfeedback format="html">',
                  u'<text>Votre réponse est incorrecte.</text>',
                u'</incorrectfeedback>',
                u'<shownumcorrect/>'
            ]
        exerciceAnswers=[]
        (bonnesReponses,mauvaisesReponses)=reponses
#        if debug: print reponses
        nombreReponses=len(bonnesReponses)+len(mauvaisesReponses)        
        scoreBonneReponse=score[len(bonnesReponses)]
        scoreMauvaiseReponse="-"+score[min(len(mauvaisesReponses),2)]
        numerosReponses=range(nombreReponses)
        random.shuffle(numerosReponses)
        for numero in numerosReponses:
          if numero<len(bonnesReponses):
            answer=[
              u'<answer fraction="%s" format="html">'%scoreBonneReponse,
                u'<text><![CDATA[<p>%s<br></p>]]></text>'%bonnesReponses[numero],
                u'<feedback format="html">',
                  u'<text></text>',
                u'</feedback>',
              u'</answer>'            
            ]
          else:
            answer=[
              u'<answer fraction="%s" format="html">'%scoreMauvaiseReponse,
                u'<text><![CDATA[<p>%s<br></p>]]></text>'%mauvaisesReponses[numero-len(bonnesReponses)],
                u'<feedback format="html">',
                  u'<text></text>',
                u'</feedback>',
              u'</answer>'            
            ]
          exerciceAnswers.extend(answer)
        exerciceTrailer=[u'</question>']
        self.forme=u"\n".join(exerciceHeader+exerciceAnswers+exerciceTrailer)
```

```python
fExercice="kanjiL1-S1-S2.txt"
cExercice=codecs.open(fExercice,"r", 'utf8')
base=cExercice.readlines()
xml=MoodleXML()
```

```python
consigneLignes=[]
nouvelExercice=False
courantExercice=False
ligneExercice=False
ligneConsigne=False
for ligne in base:
    ligne=ligne.strip()
    if u"Base de données pour un exercice" in ligne:
        if debug: print "base de données"
        nouvelExercice=True
        ligneConsigne=True
        if courantExercice:
            xml.addQuiz(mc.getMultichoices())
            courantExercice=False
    elif ligneConsigne:
        m=re.match(u"^\s*«(.*)»\s*",ligne)
        if m:
            if debug: print u"consigne trouvée", ligne
            consigne=m.group(1)
            mc=XMLMultichoice(consigne)
            ligneConsigne=False
            ligneExercice=True
    elif u"Leçon" in ligne:
        if courantExercice:
            xml.addQuiz(mc.getMultichoices())
            courantExercice=False        
        m=re.match(u"Leçon\s*(\d+)\s*",ligne)
        if m:
            mc=XMLMultichoice(m.group(1)+": "+consigne)
    elif ligneExercice and ":" in ligne:
        (kanji,choix)= ligne.split(":")
        reponses=choix.split(",")
        bonnesReponses=[]
        mauvaisesReponses=[]
        for element in reponses:
            if element.startswith("="):
                bonneReponse=element[1:]
                bonnesReponses.append(bonneReponse)
            else:
                mauvaiseReponse=element
                mauvaisesReponses.append(mauvaiseReponse)
        titre=kanji+": "+"+"+", ".join(bonnesReponses)+"-"+", ".join(mauvaisesReponses)
        mc.addExercice(MultichoiceExercice(titre, consigne, (bonnesReponses,mauvaisesReponses)))
        courantExercice=True
```
