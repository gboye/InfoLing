```python
import os,re,codecs,glob
from lxml import etree
```

```python
rep="/Users/gilles/Downloads/LNS3U5-Exercice de transcription-72051/"
```

```python
def getNom(chaine):
    nom=chaine.rsplit("/",1)[1].split("_",1)[0]
    return normalizeNom(nom)

def normalizeNom(nom):
    nomChunks=re.split(ur"-|\s",nom)
    return "".join([n.capitalize() for n in nomChunks])
    

etuFile={}
for subdir, dirs, files in os.walk(rep):
    files=[fichier for fichier in files if not fichier.startswith(".")]
    for file in files:
        print file.rsplit(".",1)[0]
        nomEtu=getNom(subdir)
        etuFile[nomEtu]=subdir+"/"+file
        print etuFile[nomEtu]
        
etuFile
```

```python
with codecs.open("/Users/gilles/Downloads/Répartition_Transcription_Sport.csv","r",encoding="utf8") as inFile:
    lines=[l.strip("\n") for l in inFile.readlines()]
    
```

```python
matchChunks={}
etuChunks={}
for line in lines[1:]:
    nom,nomUsuel,prenom,vide,sport,match,start,end=line.split("\t")
    if nomUsuel!="":
        nom=nomUsuel
    nomEtu=normalizeNom("%s %s"%(prenom,nom))
    if match!="" and nomEtu!="":
        if not match in matchChunks:
            matchChunks[match]=[]
        matchChunks[match].append(nomEtu)
        etuChunks[nomEtu]="%s-%s"%(start,end)
matchChunks
```

```python
etuChunks
```

```python
matchReps={k:k.split("_",1)[1].rsplit(".",1)[0] for k in matchChunks}
matchReps
```

```python
nonTrouves=[]
for match in matchChunks:
    print match
    for etu in matchChunks[match]:
        remis=False
        if not etu in etuFile:
            print "pb",etu,"pas de fichier"
            nonTrouves.append(etu)
        else:
            remis=True
            print etuFile[etu]
            fileType="."+etuFile[etu].rsplit(".",1)[1]
        if not etu in etuChunks or not remis:
            print "pb",etu,"pas de match"
        elif os.path.exists(etuFile[etu]):
            newFileName=rep+"%s-%s-%s"%(matchReps[match],etuChunks[etu],etu)+fileType
            print newFileName
            os.rename(etuFile[etu],newFileName)
    print
nonTrouves
```

```python
for match in ["Angleterre-Australie"]:#[matchReps[k] for k in matchReps]:
    print match
    matchLines=[]
    fichiers=glob.glob(rep+match+"*")
    for fichier in sorted([f for f in fichiers if re.search(ur"\.(trico|trs)$",f)]):
        xmlTRS=etree.parse(fichier)
        episode=xmlTRS.find("Episode")
#        print etree.tostring(episode)
        matchLines.append(etree.tostring(episode))
    with codecs.open(rep+match+".trs","w",encoding="utf8") as outFile:
        for line in matchLines:
            outFile.write(line+"\n")
    print "================"
    print
```
